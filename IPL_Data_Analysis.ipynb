```python
# No of matches per venue
sns.countplot(x='venue',data=match)
plt.xticks(rotation='vertical')
```

```python
# Matches per team

x=match['team1'].value_counts()
y=match['team2'].value_counts()
(x+y).plot(kind="barh")
```

```python
# No of the times the team won
x=pd.DataFrame({'Winner':match['winner']}).value_counts()
print(x)
```

```python
sns.countplot(x='winner',data=match)
plt.xticks(rotation='vertical')
```

```python
temp_data=match['player_of_match'].value_counts().head()
print(temp_data)
```

```python
sns.barplot(x=temp_data.index,y=temp_data.values)
plt.title('Top 5 MoM Players')
plt.xticks(rotation='vertical')
plt.xlabel('Match count')
plt.ylabel('Player')
plt.show()
```
